**Exercise 2:**

*Write a loop to generate 100 samples (of 100 measurements each) from the same normal distribution. Save these sample means in an array.*

We want all of our samples to come from the same distribution, so let's generate a population to sample from.

```python
%%R
# Generate a normally distributed population (N=1000) with a mean of 247 and an standard deviation of 19
population = rnorm(1000, mean=247, sd=19)
```

Now that we have our population, we can sample from it.

```python
%%R
# From our population, take random 100 samples, each with n = 100, and write the mean of each sample to sample_means.
sample_means = sapply(seq(1:100), function(x) mean(sample(population, size=100, replace=FALSE)))
```

**Exercise 3:**

Because IPython keeps everything in-line, we don't need to create a new window.

```python
%%R
# Plot a histogram of our sample means.
hist(sample_means)
# Plot a dotted line marking the mean of sample means.
abline(v=mean(sample_means), lty=2)
```

```python
%%R
# Plot a histogram of the actual population to compare to our DOSM.
hist(population)
# As above, mark the mean with a dotted line.
abline(v=mean(population), lty=2)
```

**Exercise 4:** 

*Calculate the standard deviation of the DOSM and compare to the standard deviation of the population.*

```python
%%R
# Calculate the SD of our DOSM
dosm_sd = sd(sample_means)
# For comparison, calculate the SD of the entire population.
pop_sd = sd(population)
# Print the SD values.
cat(" DOSM SD =", dosm_sd, "\n", "Population SD =", pop_sd)
```

As expected, the SD of the DOSM is much smaller than that of the actual population.

**Exercise 5:** 

*Use the arrows command to draw an arrow showing the SEM*

First, we need to calculate the SEM for the DOSM.

```python
%%R
# Calculate the SEM by dividing the DOSM SD by the square root of the sample size for each sample.
dosm_sem = dosm_sd / sqrt(100)
# Print out the DOSM SEM so we can see what the value is.
cat(" DOSM SEM =", dosm_sem)
```

```python
%%R
# Plot a histogram of our sample means.
hist(sample_means)
# Use the arrows command to draw a line from -1 SEM to +1 SEM. Draw the line at y=2 because it looks nice there.
# Turn off the actual arrows by setting code=3 (put arrows on both ends) and length=0 (arrow length of 0).
arrows(mean(sample_means)-dosm_sem, 2, mean(sample_means)+dosm_sem, 2, code=3, length=0)
# Because SEM is related to the mean, let's draw a dotted line at the mean as well.
abline(v=mean(sample_means), lty=2)
```

**Exercises 6 and 7:**

*Calculate the sample mean for samples of 10, 20, 50, 100, 200, 500, and 1000. Show how to do this using both a standard 'for' loop as well as the 'apply' function.*

First, we're going to define a custom function to calculate the mean, SD, and 95% CI for a random sample of size N from a specified distribution. 

(We will need the 95% CI in Exercise 8, and it is most efficient to calculate everything in a single function)

```python
%%R
# Define a new function called genSampleMean with two arguments, sample_size and distribution.
genSamples = function(sample_size, distribution) {
    # Take a sample of the specified size from the specified distribution.
    sample = sample(population, size=sample_size, replace=FALSE)
    # Get the mean of our sample.
    sample_mean = mean(sample)
    # Get the SD for our sample.
    sample_sd = sd(sample)
    # Calculate the 95% confidence interval for our sample mean.
    sample_error = (qnorm(0.975)*sample_sd)/(sqrt(sample_size))
    sample_ci_ub = sample_mean + sample_error
    sample_ci_lb = sample_mean - sample_error
    # Group the mean, SD, and error into a vector.
    sample_stats = c(sample_size, sample_mean, sample_sd, sample_error, sample_ci_ub, sample_ci_lb)
    # Output our vector of statistics.
    return(sample_stats)
}
```

Loops in R are slow, so let's use the apply function.

```python
%%R
# Create a vector containing all of the different sample sizes we want.
sample_sizes = cbind(10, 20, 50, 100, 200, 500, 1000)
# Generate statistics for each of sample sizes in our vector, and save it to a new array called multi_sample.
multi_sample = sapply(sample_sizes, function(x) genSamples(x, population))
```

Here's how we would do the same thing using a 'for' loop.


```python
%%R
# Create a new empty array to hold our new sample means.
loop_multi_sample = c()
# For every sample size in our list...
for (i in sample_sizes) {
    # Calculate statistics for a sample of the given size.
    loop_sample_stats = genSamples(i, population)
    # Add this mean value to our array by re-defining it as the existing array + the new vector of statistics.
    loop_multi_sample = c(loop_multi_sample, loop_sample_stats)
}
```

The statistics we calculated for each sample size were dumped into a 2D array, but data frames are easier to work with, so lets convert our array to a data frame.

```python
%%R
# Read the array as a data frame.
multi_sample_df = as.data.frame(multi_sample)
# Transpose the rows and columns in our new data frame.
# We do this because our array was structured with sample sizes as columns instead of rows.
multi_sample_df = data.frame(t(multi_sample_df))
# Print out our lovely new data frame.
multi_sample_df
```

Now we can create the plot. We're going to use ggplot2 because it lets us plot the CI's easily.

```python
%%R
# Load ggplot2.
library(ggplot2)
# Create a new plot from multi_sample_df, with X1 (sample size) on the x-axis and X2 (sample mean) on the y-axis
ggplot(multi_sample_df, aes(x=X1, y=X2)) + 
    # At each point, draw error bars based on the upper (X5) and lower (X6) bounds of the CI.
    geom_errorbar(aes(ymin=X6, ymax=X5), width=25, alpha=0.5) +
    # Plot the data points, not just the error bars.
    geom_point() +
    # Draw a horizontal dashed line at the mean of means.
    geom_hline(yintercept=mean(sample_means), linetype="dashed") +
    # Label our axes.
    xlab("Sample Mean") +
    ylab("Sample Size")
```
